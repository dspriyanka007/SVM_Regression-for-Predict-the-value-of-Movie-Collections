### Predict the value of collection by giving the values of other 17 variables by using regression model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Movie_regression.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

### missing value imputation

```python
df.Time_taken=df.Time_taken.fillna(df.Time_taken.mean())
```

```python
df.info()
```

### Dummy variable creation

```python
df=pd.get_dummies(df)
```

```python
df.head()
```

```python
del df['Genre_Action']
```

```python
df.head()
```

### x-y split

```python
x=df.loc[:,df.columns!='Collection']
```

```python
y=df['Collection']
```

```python
x.head()
```

```python
x.shape
```

```python
y.shape
```

### Train-Test-Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.head()
```

```python
y_train.shape
```

```python
y_test.shape
```

### Standardizing Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler().fit(x_train)
```

```python
x_train_std=sc.transform(x_train)
```

```python
x_test_std=sc.transform(x_test)
```

```python
x_test_std
```

```python
x_test
```

### Training Regression Tree

```python
from sklearn.svm import SVR
```

```python
svr=SVR(kernel='linear',C=5000)
```

```python
svr.fit(x_train_std,y_train)
```

### Predict value using trained model

```python
y_test_pred=svr.predict(x_test_std)
y_train_pred=svr.predict(x_train_std)
```

```python
y_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```
